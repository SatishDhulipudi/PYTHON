## Exploratory Data Analysis(EDA)

### Gender

### plotting a bar chart for Gender and count

```python
ax = sns.countplot(x = 'Gender',data = data)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

### plotting a bar chart for gender and total amount

```python
ax = data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount')
sns.barplot(x='Gender',y='Amount',data = ax)
plt.show()
```

#### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

### Age

```python
ax = sns.countplot(data = data,x = 'Age Group',hue ='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

### Total Amount vs Group

```python
ax = data.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount')
sns.barplot(x = 'Age Group',y = 'Amount',data = ax)
plt.show()
```

#### From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

### State

#### Total number of orders from top 10 states

```python
ax = data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = ax, x = 'State',y= 'Orders')
```

#### Total amount by sales from top 10 states

```python
ax = data.groupby(['State'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.barplot(data = ax ,x = 'State',y = 'Amount')
```

#### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

### Marital Status

```python
ax = sns.countplot(data = data,x = 'Marital_Status')
sns.set(rc={'figure.figsize':(7,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = data.groupby(['Marital_Status','Gender'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(data= ax,x='Marital_Status',y = 'Amount',hue = 'Gender')
sns.set(rc={'figure.figsize':(6,5)})
plt.show()
```

#### From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

### Occupation

```python
sns.set(rc={'figure.figsize':(20,10)})
ax = sns.countplot(data = data,x= 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
ax = data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
sns.barplot(data = ax,x = 'Occupation',y ='Amount')
plt.show()
```

#### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

### Product Category

```python
ax = sns.countplot(data = data,x ='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
ax = data.groupby(['Product_Category'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending=False).head(10)
sns.barplot(data = ax,x = 'Product_Category',y = 'Amount')
plt.show()
```

#### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
ax = data.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.barplot(data = ax,x ='Product_ID', y = 'Orders')
sns.set(rc={'figure.figsize':(20,10)})
plt.show()
```

### Top 10 most sold products

```python
ax = plt.subplots(figsize=(12,7))
ax1 = data.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
plt.show()
```

### Conclusion

- Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
